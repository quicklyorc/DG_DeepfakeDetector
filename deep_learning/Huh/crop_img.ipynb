```python
import cv2
import pandas as pd
import numpy as np
import glob
from google.colab.patches import cv2_imshow
import mediapipe as mp
from mediapipe.tasks import python
from mediapipe.tasks.python import vision

f_num = 1

imct = glob.glob(f"/content/drive/MyDrive/crop/Real_img{f_num}/*")
print(len(imct))

imct = glob.glob(f"/content/drive/MyDrive/crop/Fake_img{f_num}/*")
print(len(imct))

imct = glob.glob(f"/content/drive/MyDrive/aihub/딥페이크 변조 영상/train_img/Real_img/Real_img{f_num}/*")
print(len(imct))

imct = glob.glob(f"/content/drive/MyDrive/aihub/딥페이크 변조 영상/train_img/Fake_img/Fake_img{f_num}/*")
print(len(imct))
```

```python

# 이미지 로드 및 라벨링
# 원본 폴더에서 사진을 읽어와서 전처리
fake_df = pd.read_csv(f"/content/drive/MyDrive/aihub/딥페이크 변조 영상/train_img/Fake_img/Fake_list{f_num}.csv")
real_df = pd.read_csv(f"/content/drive/MyDrive/aihub/딥페이크 변조 영상/train_img/Real_img/Real_list{f_num}.csv")
fake_list = list(fake_df["0"])
real_list = list(real_df["0"])

print(len(real_list))

for idx, img_name in enumerate(real_list):
    #img=cv2.imread(img_name)
    img = img_name
    img_cv = cv2.imread(img_name)

    # STEP 2: Create an FaceDetector object.
    base_options = python.BaseOptions(model_asset_path='/content/drive/MyDrive/detector.tflite')
    options = vision.FaceDetectorOptions(base_options=base_options)
    detector = vision.FaceDetector.create_from_options(options)

    # STEP 3: Load the input image.
    image = mp.Image.create_from_file(img)

    # STEP 4: Detect faces in the input image.
    detection_result = detector.detect(image)

    x = int(detection_result.detections[0].bounding_box.origin_x)
    y = int(detection_result.detections[0].bounding_box.origin_y)
    w = int(detection_result.detections[0].bounding_box.width)
    h = int(detection_result.detections[0].bounding_box.height)

    roi = img_cv[y:y+h, x:x+w]
    resized_roi = cv2.resize(roi,(256, 256))

    name = img_name.split("/")[-1]
    save_file_path = f"/content/drive/MyDrive/crop/Real_img{f_num}/{name}_crop.jpg"
    cv2.imwrite(save_file_path, roi)
    print(name, "완료", idx)
```

```python
import pandas as pd
import numpy as np
import glob
import zipfile
import random
import os
import shutil

def unzip(file, path):
    with zipfile.ZipFile(file, 'r') as zip_ref:
        zip_ref.extractall(path)

# 본인 이름 적기
worker = "동우"

work_dict = {"동우":(0,10), "유리":(10,19), "가연":(19,28), "대선":(28,38), "지윤":(38,47)}
download_path = "/content/drive/MyDrive/aihub/딥페이크 변조 영상/1.Training"
zip_list = glob.glob(download_path+"/*")
zip_list = zip_list[work_dict[worker][0]: work_dict[worker][1]]
zip_list
```

```python
zip_path.split("/")[-1][16] == 'ᄋ'
```

```python
for zip_path in zip_list:

    # 압축 풀 위치
    unzip_path = "/content/drive/MyDrive/aihub/딥페이크 변조 영상/tmp"

    unzip(zip_path, unzip_path)

    # real_path => 원본영상 저장할 경로
    # fake_path => 딥페이크 영상 저장할 경로

    # 경로간 구분은 "/"로 하고, 절대경로를 사용할 것!!!
    real_path = "/content/drive/MyDrive/aihub/딥페이크 변조 영상/Real"
    fake_path = "/content/drive/MyDrive/aihub/딥페이크 변조 영상/Fake"

    # 영상 라벨링 및 이동
    if zip_path.split("/")[-1][15] == 'a':

        # audio_driven 딥페이크 동영상의 경우

        uuid = glob.glob(unzip_path+"/"+zip_path.split("/")[-1].split("]")[2].split(".")[0]+"/*")
        print(uuid)

        for i in uuid:
            videos = glob.glob(i+"/*")
            if len(videos) >=30:
                rd_videos = random.sample(videos, 30)
            else:
                rd_videos= videos

            for j in rd_videos:

                shutil.move(j, fake_path+"/"+j.split("/")[-1])

    elif zip_path.split("/")[-1][15] == 't':

        # audio_driven외의 딥페이크 동영상의 경우

        uuid = glob.glob(unzip_path+"/"+zip_path.split("/")[-1].split("_")[1]+"/*")
        print(uuid)

        for i in uuid:
            videos = glob.glob(i+"/*")
            if len(videos) >=30:
                rd_videos = random.sample(videos, 30)
            else:
                rd_videos= videos

            for j in rd_videos:

                shutil.move(j, fake_path+"/"+j.split("/")[-1])

    elif zip_path.split("/")[-1][16] == 'ᄋ':

        # 원본 동영상(진짜)인 경우

        uuid = glob.glob(unzip_path+"/"+zip_path.split("/")[-1].split("]")[2].split(".")[0]+"/*")
        print(uuid)

        for i in uuid:
            videos = glob.glob(i+"/*")

            for j in videos:

                shutil.move(j, real_path+"/"+j.split("/")[-1])

    else:
        print("잘못된 파일임!")
```

```python
# 가짜영상 확인
mp4_list = glob.glob("/content/drive/MyDrive/aihub/딥페이크 변조 영상/Fake/*")
print(len(mp4_list))
df = pd.read_csv("/content/drive/MyDrive/aihub/변조영상_training_메타데이터.csv")

ff = []
for i in range(len(mp4_list)):
  ff.append(mp4_list[i].split('/')[-1])
len(ff) # Fake 영상 ID 모음

ff_id = pd.DataFrame(ff, columns=['변조ID'])

df_id = df['영상ID'] # aihub

overlap_count = pd.merge(df_id, ff_id, left_on='영상ID', right_on='변조ID', how='inner').shape[0]
print("중복되는 값의 개수:", overlap_count)
```

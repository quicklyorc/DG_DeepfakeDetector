```python
import pandas as pd
import numpy as np
import glob
import zipfile
import random
import os
import shutil

def unzip(file, path):
    with zipfile.ZipFile(file, 'r') as zip_ref:
        zip_ref.extractall(path)

# 본인 이름 적기
worker = "가연"

work_dict = {"동우":(0,10), "유리":(10,19), "가연":(19,28), "대선":(28,38), "지윤":(38,47)}
download_path = "/content/drive/MyDrive/aihub/딥페이크 변조 영상/1.Training"
zip_list = glob.glob(download_path+"/*")
zip_list = zip_list[work_dict[worker][0]: work_dict[worker][1]]
zip_list
```

```python
for zip_path in zip_list:

    # 압축 풀 위치
    unzip_path = "/content/drive/MyDrive/aihub/딥페이크 변조 영상/tmp"

    # 라벨링 csv파일 저장 위치
    # 우리가 새로 만들 라벨링 csv파일 저장 위치임.
    csv_path = "/content/drive/MyDrive/aihub/딥페이크 변조 영상"

    if not os.path.exists(csv_path):
        os.makedirs(csv_path)

    if os.path.exists(csv_path+"/label_r.csv"):
        label_df = pd.read_csv(csv_path+"/label_r.csv")
    else:
        label_df = pd.DataFrame(columns=["video","Real_or_Fake"])

    unzip(zip_path, unzip_path)

    # real_path => 원본영상 저장할 경로
    # fake_path => 딥페이크 영상 저장할 경로

    # 경로간 구분은 "/"로 하고, 절대경로를 사용할 것!!!
    real_path = "/content/drive/MyDrive/aihub/딥페이크 변조 영상/Real"
    fake_path = "/content/drive/MyDrive/aihub/딥페이크 변조 영상/Fake"

    # 대상 디렉토리가 존재하지 않으면 생성
    if not os.path.exists(real_path):
        os.makedirs(real_path)

    if not os.path.exists(fake_path):
        os.makedirs(fake_path)

    # 영상 라벨링 및 이동
    if zip_path.split("/")[-1][15] == 'a':

        # audio_driven 딥페이크 동영상의 경우

        uuid = glob.glob(unzip_path+"/"+zip_path.split("/")[-1].split("]")[2].split(".")[0]+"/*")
        print(uuid)

        for i in uuid:
            videos = glob.glob(i+"/*")
            if len(videos) >=30:
                rd_videos = random.sample(videos, 30)
            else:
                rd_videos= videos
            for j in rd_videos:

                n_row = pd.DataFrame({"video": [fake_path+"/"+j.split("/")[-1]], "Real_or_Fake": [False]})
                label_df = pd.concat((label_df,n_row), ignore_index=True)

                shutil.move(j, fake_path+"/"+j.split("/")[-1])

    elif zip_path.split("/")[-1][15] == 't':

        # audio_driven외의 딥페이크 동영상의 경우

        uuid = glob.glob(unzip_path+"/"+zip_path.split("/")[-1].split("_")[1]+"/*")
        print(uuid)

        for i in uuid:
            videos = glob.glob(i+"/*")
            if len(videos) >=30:
                rd_videos = random.sample(videos, 30)
            else:
                rd_videos= videos
            for j in rd_videos:

                n_row = pd.DataFrame({"video": [fake_path+"/"+j.split("/")[-1]], "Real_or_Fake": [False]})
                label_df = pd.concat((label_df,n_row), ignore_index=True)

                shutil.move(j, fake_path+"/"+j.split("/")[-1])

    elif zip_path.split("/")[-1][16] == "ᄋ":

        # 원본 동영상(진짜)인 경우

        uuid = glob.glob(unzip_path+"/"+zip_path.split("/")[-1].split("]")[2].split(".")[0]+"/*")
        print(uuid)

        for i in uuid:
            videos = glob.glob(i+"/*")
            for j in videos:

                n_row = pd.DataFrame({"video": [real_path+"/"+j.split("/")[-1]], "Real_or_Fake": [True]})
                label_df = pd.concat((label_df,n_row), ignore_index=True)

                shutil.move(j, real_path+"/"+j.split("/")[-1])

    else:
        print("잘못된 파일임!")
```

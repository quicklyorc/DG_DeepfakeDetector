```python
import pandas as pd
import numpy as np
import glob
```

```python
mp4_list = glob.glob("/content/drive/MyDrive/aihub/딥페이크 변조 영상/Fake/*")
print(len(mp4_list))
df = pd.read_csv("/content/drive/MyDrive/aihub/변조영상_training_메타데이터.csv")

ff = []
for i in range(len(mp4_list)):
  ff.append(mp4_list[i].split('/')[-1])
len(ff) # Fake 영상 ID 모음

ff_id = pd.DataFrame(ff, columns=['변조ID'])

df_id = df['영상ID'] # aihub

overlap_count = pd.merge(df_id, ff_id, left_on='영상ID', right_on='변조ID', how='inner').shape[0]
print("중복되는 값의 개수:", overlap_count)
```

```python
mp4_list = glob.glob("/content/drive/MyDrive/aihub/딥페이크 변조 영상/Real/*")
print(len(mp4_list))
df = pd.read_csv("/content/drive/MyDrive/aihub/원본영상_training_메타데이터.csv")

ff = []
for i in range(len(mp4_list)):
  ff.append(mp4_list[i].split('/')[-1])
len(ff) # Fake 영상 ID 모음

ff_id = pd.DataFrame(ff, columns=['변조ID'])

df_id = df['영상ID'] # aihub

overlap_count = pd.merge(df_id, ff_id, left_on='영상ID', right_on='변조ID', how='inner').shape[0]
print("중복되는 값의 개수:", overlap_count)
```

### train 영상 label

```python
fake_path = "/content/drive/MyDrive/aihub/딥페이크 변조 영상/Fake"
real_path = "/content/drive/MyDrive/aihub/딥페이크 변조 영상/Real"

fake_videos = glob.glob(fake_path+"/*")
real_videos = glob.glob(real_path+"/*")
```

```python
len(fake_videos)
```

```python
len(real_videos)
```

```python
label_fake = pd.DataFrame(columns = ["video","label"])
label_fake["video"] = fake_videos
label_fake["label"] = False
label_real = pd.DataFrame(columns = ["video","label"])
label_real["video"] = real_videos
label_real["label"] = True
```

```python
label_df = pd.concat((label_fake, label_real), ignore_index=True)
label_df.tail()
```

```python
label_df.head()
```

```python
label_df.to_csv("/content/drive/MyDrive/aihub/딥페이크 변조 영상/label.csv")
```
